```python
from pathlib import Path
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = Path('M2_monthly_data_clean.csv')
MS_df = pd.read_csv(data)
MS_df['Monthly '] = pd.to_datetime(MS_df['Monthly '], format='%m/%d/%Y')
MS_df.rename(columns={"Monthly ": "Date"}, inplace= True)
MS_df.head()
```

```python
data = Path('BTC-USD (2014-2024)_2.csv')
Bit_df = pd.read_csv(data)
Bit_df['Date'] = pd.to_datetime(Bit_df['Date'], format='%m/%d/%Y')
Bit_df.head()
```

```python

# Convert the 'Dates' column in Bit_df to datetime if not already
Bit_df['Date'] = pd.to_datetime(Bit_df['Date'])

monthly_Bit = Bit_df[Bit_df['Date'].isin(MS_df['Date'])]

monthly_Bit
```

```python
def calculate_growth_rate(df, column_name):
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    growth_rate = df[column_name].pct_change() * 100
    return growth_rate
```

```python
monthly_Bit['Growth_rate'] = calculate_growth_rate(monthly_Bit, 'Close')

monthly_Bit
```

```python
MS_df['Growth_rate'] = MS_df['Growth_rate']* 100
MS_df = MS_df.rename(columns={"Monthly": "Date"})
MS_df_clean = MS_df[MS_df['Date'].isin(monthly_Bit['Date'])]
MS_df_clean
```

```python
merged_df = pd.merge(monthly_Bit, MS_df, on="Date", how='inner')

merged_df.head()
```

```python
def classify_columns(row):

    if (row['Growth_rate_x'] < 0 and row['Growth_rate_y'] < 0) or (row['Growth_rate_x'] > 0 and row['Growth_rate_y'] > 0):
        return 1
    else:
        return 0
```

```python
merged_df['Trend'] = merged_df.apply(classify_columns, axis=1)
merged_df
```

```python
merged_df.to_csv('Bit_MS.csv', index=False)
```
